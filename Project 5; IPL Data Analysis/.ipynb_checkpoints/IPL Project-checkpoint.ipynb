```python
matches_played = match_data['team1'].value_counts() + match_data['team2'].value_counts()
```

```python
type(matches_played)
```

```python
matches_played_df = matches_played.to_frame().reset_index()
```

```python
matches_played_df.columns = ['team_name', 'Matches_played']
```

```python
matches_played_df
```

```python
wins = pd.DataFrame(match_data['winner'].value_counts()).reset_index()
wins
```

```python
# If the columns are named 'team_name' in matches_played_df and 'team' in wins
played = matches_played_df.merge(wins, left_on='team_name', right_on='winner', how='inner')
played
```

```python
 matches_played_df.columns
```

```python
wins.columns
```

```python
trace1 = go.Bar(
    x = played['team_name'],
    y = played['Matches_played'],
    name = 'Total Matches'
)

trace2 = go.Bar(
    x = played['team_name'],
    y = played['count'],
    name = 'Matches won'
) 
```

```python
data = [trace1, trace2]
```

```python
iplot(data)
```
